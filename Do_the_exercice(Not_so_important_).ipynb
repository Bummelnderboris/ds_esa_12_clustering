```python
# This notebooks solves the given exercice but is ment to be as the second option
# But at the moment i am trying to finish the k means algorithm from scratch and hope to be done before the review
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

df= pd.read_csv('https://drive.google.com/uc?id=1sTO90PQRtbK80XdzMJ2KttRRtt9M7xuq')
ds = df.to_numpy()
```

```python
dist = []
c = range(1,15)
for k in c:
  kmeanModel = KMeans(n_clusters=k)
  kmeanModel.fit(df)
  dist.append(kmeanModel.inertia_)

plt.plot(c, dist, 'bx-')
```

```python
# It seems like the optimal k is 5 or 6. Let's go with 5

model = KMeans(n_clusters=5)
model.fit(df)
labels = model.predict(df)

#df['pred_class'] = labels
df[0:10]
```

```python
#B: Plot the clusteres in a distinguishable form
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(ds[:, 0], ds[:, 1], ds[:, 2], c=labels)
```
